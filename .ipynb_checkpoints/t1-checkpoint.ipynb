```python
data = pd.read_csv('1/Demdin(1).csv', sep=';', index_col=0)
data.plot(subplots=True, layout=(5,1), figsize=(5,10))
data.tail()
```

```python
y = data.M1
x = data[['ipc', 'rc', 'y']]
x = sm.add_constant(x)
```

```python
modelo = sm.OLS(endog=y, exog=x)
reg = modelo.fit()
reg.summary()
```

```python
logdata = np.log(data)
```

```python
x = logdata[['ipc', 'rc', 'y']]
x = sm.add_constant(x)
y = logdata.M1
```

```python
modelo = sm.OLS(endog=y, exog=x)
reg = modelo.fit()
reg.summary()
```

```python
logdata.plot(subplots=True, layout=(5,1), figsize=(5,10))
```

```python
# grafico de los residuos
reg.resid.plot()
```
